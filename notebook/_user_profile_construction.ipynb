```python
import pandas as pd 
pnews_df  = pd.read_csv("../data/pure_news.csv")
pnews_df.head()
```

```python
behavior_df = pd.read_csv("../data/clean_behaviors.csv")
behavior_df.head()
```

```python
import random
# تستخدم لتثبيت القيمه العشوائيه
random.seed(1)
# يقوم باختيار رقم معين بطريقه عشوائيه
user_id = random.choice(behavior_df["user_id"])
# يختار مصفوفه معينه بناءا علي اليوزر المختار و يقوم باستخراج سجل تاريخه و يقوم بعمل مصفوفه و وصعه فيه 
user_history = behavior_df.loc[behavior_df["user_id"] == user_id, "history"].iloc[0]
# تقوم بفصل كل كلمه عن الاخري
read_article_ids = user_history.split()
user_id
```

```python
# يقوم بتصفيه المقالات علي حسب تفضيلات اليوزر
user_articles =  pnews_df[pnews_df['newsid'].isin(read_article_ids)]
# يقوم بتنظيف النص و تطبيق علي بعض الاوامر المخزنه
user_articles_cleaned = " ".join(user_articles["pure_content"])
# سنقوم بتسجيل ملاحظات العملاء عن المقالات
with open ("../result/feedback_users.txt","w") as f:
    f.write(user_articles_cleaned)
```
